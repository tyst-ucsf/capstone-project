```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the spiral
theta = np.linspace(0, 10 * np.pi, 1000)  # Angle from 0 to 10
r = np.linspace(0, 10, 1000)  # Radius increases linearly

# Convert polar coordinates to Cartesian coordinates
x = r * np.cos(theta)
y = r * np.sin(theta)

# Plot the spiral
plt.figure(figsize=(8, 8))
plt.plot(x, y, label='Spiral', linewidth=2)
plt.title('Spiral Graph', fontsize=16)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.grid(False)
plt.axis('equal')  # Ensure the scale is equal
plt.legend()
plt.show()
```
